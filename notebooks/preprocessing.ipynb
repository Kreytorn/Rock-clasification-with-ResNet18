```python
import os
from PIL import Image, ImageFilter
from tqdm import tqdm
import random

# === Your root dataset path
BASE_PATH = r"C:\Users\kuzey\OneDrive\Masaüstü\cografya_proje\Dataset"

# === Target resolution
TARGET_SIZE = (224, 224)

# === Loop and clean every image
for main_class in os.listdir(BASE_PATH):
    main_path = os.path.join(BASE_PATH, main_class)
    if not os.path.isdir(main_path): continue

    for subclass in os.listdir(main_path):
        sub_path = os.path.join(main_path, subclass)
        if not os.path.isdir(sub_path): continue

        print(f"Preprocessing: {main_class}/{subclass}")
        for fname in tqdm(os.listdir(sub_path)):
            if not fname.lower().endswith((".jpg", ".jpeg", ".png")): continue

            img_path = os.path.join(sub_path, fname)
            try:
                img = Image.open(img_path).convert("RGB")

                # Resize
                img = img.resize(TARGET_SIZE)

                # Optional: slight sharpening for texture clarity
                if random.random() < 0.25:
                    img = img.filter(ImageFilter.SHARPEN)

                # Save as clean JPG
                img.save(img_path, format="JPEG", quality=95)

            except Exception as e:
                print(f"⚠️ Skipping {fname}: {e}")
```

```python
import os
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# === SET YOUR DATASET PATH HERE ===
DATASET_PATH = r"C:\Users\kuzey\OneDrive\Masaüstü\cografya_proje\Dataset"  # <- Update if needed

brightness_per_image = []
class_brightness = {}

# Loop through each subclass folder
for subclass in os.listdir(DATASET_PATH):
    subclass_path = os.path.join(DATASET_PATH, subclass)
    if not os.path.isdir(subclass_path):
        continue

    subclass_brightness = []

    for fname in os.listdir(subclass_path):
        if fname.lower().endswith(".jpg"):
            img_path = os.path.join(subclass_path, fname)
            try:
                img = Image.open(img_path).convert("L")  # grayscale
                img_np = np.asarray(img) / 255.0  # normalize to [0, 1]
                brightness = img_np.mean()
                brightness_per_image.append(brightness)
                subclass_brightness.append(brightness)
            except:
                continue

    if subclass_brightness:
        class_brightness[subclass] = subclass_brightness

# === Plot average brightness per class
avg_brightness = {cls: np.mean(vals) for cls, vals in class_brightness.items()}
plt.figure(figsize=(10, 5))
plt.bar(avg_brightness.keys(), avg_brightness.values(), color='skyblue')
plt.ylabel("Average Brightness (0-1)")
plt.title("Average Brightness per Rock Class")
plt.xticks(rotation=30)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# === Histogram of all image brightness
plt.figure(figsize=(10, 5))
plt.hist(brightness_per_image, bins=50, color='gray', edgecolor='black')
plt.title("Distribution of Image Brightness Across Dataset")
plt.xlabel("Brightness (0 = dark, 1 = bright)")
plt.ylabel("Number of Images")
plt.grid(True)
plt.tight_layout()
plt.show()

# === Print stats
brightest = max(brightness_per_image)
darkest = min(brightness_per_image)
avg_total = np.mean(brightness_per_image)

print("🌞 Brightest Image Brightness:", round(brightest, 4))
print("🌑 Darkest Image Brightness:", round(darkest, 4))
print("📊 Average Brightness of Dataset:", round(avg_total, 4))
```

```python
import os
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from skimage.filters import laplace

# === Set your actual dataset path here ===
DATASET_PATH = r"C:\Users\kuzey\OneDrive\Masaüstü\cografya_proje\Dataset"

data = []

# === Process each image in each class
for subclass in os.listdir(DATASET_PATH):
    subclass_path = os.path.join(DATASET_PATH, subclass)
    if not os.path.isdir(subclass_path):
        continue

    for fname in os.listdir(subclass_path):
        if not fname.lower().endswith(".jpg"):
            continue

        img_path = os.path.join(subclass_path, fname)

        try:
            img = Image.open(img_path).convert("RGB")
            img_gray = img.convert("L")

            img_np = np.array(img_gray) / 255.0
            img_rgb = np.array(img) / 255.0

            brightness = img_np.mean()
            contrast = img_np.std()
            sharpness = laplace(img_np).var()  # Edge detail
            r_mean = img_rgb[:, :, 0].mean()
            g_mean = img_rgb[:, :, 1].mean()
            b_mean = img_rgb[:, :, 2].mean()

            data.append({
                "class": subclass,
                "brightness": brightness,
                "contrast": contrast,
                "sharpness": sharpness,
                "r_mean": r_mean,
                "g_mean": g_mean,
                "b_mean": b_mean
            })
        except:
            continue

# === Convert to DataFrame
df = pd.DataFrame(data)

# === Plot 1: Average Brightness per Class
plt.figure(figsize=(10, 5))
df.groupby("class")["brightness"].mean().sort_values().plot(kind='bar', color='orange')
plt.title("🔆 Average Brightness per Class")
plt.ylabel("Brightness (0 = dark, 1 = bright)")
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# === Plot 2: Contrast per Class (Boxplot)
plt.figure(figsize=(10, 5))
df.boxplot(column="contrast", by="class")
plt.title("🖤 Contrast Distribution per Class")
plt.suptitle("")
plt.ylabel("Pixel Intensity Std Dev")
plt.tight_layout()
plt.show()

# === Plot 3: Sharpness per Class (Boxplot)
plt.figure(figsize=(10, 5))
df.boxplot(column="sharpness", by="class")
plt.title("🔪 Sharpness Distribution per Class")
plt.suptitle("")
plt.ylabel("Laplacian Variance")
plt.tight_layout()
plt.show()

# === Plot 4: RGB Mean per Class (Stacked Bar)
rgb_means = df.groupby("class")[["r_mean", "g_mean", "b_mean"]].mean()
rgb_means.plot(kind='bar', stacked=True, figsize=(10, 5), colormap='viridis')
plt.title("🎨 Average R/G/B Intensity per Class")
plt.ylabel("Normalized Channel Intensity")
plt.tight_layout()
plt.show()
```
